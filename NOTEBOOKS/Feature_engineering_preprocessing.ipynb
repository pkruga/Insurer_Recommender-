#### **Label Encoding**

```python
label_encoder = LabelEncoder()
df['Insurer_Encoded'] = label_encoder.fit_transform(df['Insurer'])
df.head()
```

#### **Creation of Quarter and Year Column**

```python
# Create a quarter column from the date

df['Date'] = pd.to_datetime(df['Date'])
df['Quarter'] = df['Date'].dt.quarter
df.head()
```

```python
# prompt: use the previous code to create a year column

df['Year'] = df['Date'].dt.year
df.head()
```

#### **Reliability Score**

We are going to create a reliability score from the ratios columns.


```python
# Create a column that will calculate the reliability score for insurer depending on claims payment ratio, claims declined ratio, and claims closed as no claims

# Define weights for each factor (adjust these as needed)
weight_payment_ratio = 0.6
weight_declined_ratio = 0.2
weight_no_claims_ratio = 0.2

# Calculate the reliability score
df['Reliability_Score'] = (
    (df['Claim_payment_ratio_(%)'] / 100) * weight_payment_ratio
    - (df['Claims_declined_ratio_(%)'] / 100) * weight_declined_ratio
    - (df['Claims_closed_as_no_claims_ratio_(%)'] / 100) * weight_no_claims_ratio
)

# Scale the score to a range of 0-100 (optional)
scaler = MinMaxScaler(feature_range=(0, 100))
df['Reliability_Score'] = scaler.fit_transform(df[['Reliability_Score']])

df.head()
```

#### **Label encoding of our Reliability Score**

```python
# Label encode our reliability score based on the figures given from poor to very good

# Define the bins and labels
bins = [0, 20, 40, 60, 80, 100]
labels = ['Very Poor', 'Poor', 'Average', 'Good', 'Very Good']

# Use pd.cut to discretize the 'Reliability_Score' column
df['Reliability_Label'] = pd.cut(df['Reliability_Score'], bins=bins, labels=labels, right=False)

# Display the updated DataFrame
df.head()
```

### **Multicollinearity assessment using VIF**

```python
# Identify numerical columns
numerical_cols = df.select_dtypes(include=['float', 'int']).columns

# Print the numerical columns
print(numerical_cols)
```

```python
# Perform a variance inflation factor for the relevant columns

# Calculate VIF for each numerical column
vif_data = pd.DataFrame()
vif_data["Variable"] = numerical_cols
vif_data["VIF"] = [variance_inflation_factor(df[numerical_cols].values, i) for i in range(len(numerical_cols))]

# Print the VIF results
print(vif_data)
```

```python
# Create a report based on our VIF results from the previous code

# Interpret the VIF results
print("\nInterpretation of VIF Results:\n")

# Identify high VIF values
high_vif_vars = vif_data[vif_data["VIF"] > 5]["Variable"].tolist()
if high_vif_vars:
    print(f"- The following variables exhibit high multicollinearity (VIF > 5): {', '.join(high_vif_vars)}")
else:
    print("- No variables show concerning levels of multicollinearity (VIF < 5).")

# Explain implications
print("\nKey Implications:")
if high_vif_vars:
    print("- High multicollinearity can lead to unstable model coefficients and inflated standard errors.")
    print("- It might be necessary to address multicollinearity by removing or transforming some of the high VIF variables.")
else:
    print("- The absence of high VIF values suggests that multicollinearity is not a major concern for the current set of features.")

print("\nFurther Analysis:")
if high_vif_vars:
    print("- Consider techniques like feature selection, dimensionality reduction (PCA), or regularization (Ridge, Lasso) to mitigate multicollinearity.")
    print("- Examine the correlation matrix to identify specific pairs of variables with high correlations.")
print("- Evaluate the impact of addressing multicollinearity on model performance and interpretability.")
```

```python
# Dropping columns with unecessary columns factoring VIF
X = df.drop(['Claims_outstanding_at_the_beginning_of_the_quarter','Claims_intimated_during_the_quarter','Claims_revived','Claims_paid','Claims_declined','Claims_outstanding_at_the_end','Total_Claims_Payable','Total_Claims_Action_during_the_Quarter', 'Claims_closed_as_no_claims_ratio_(%)'], axis=1)

# Include object type columns
X = pd.concat([X, df[['Date', 'Insurer']]], axis=1)

print(X.head())
```
